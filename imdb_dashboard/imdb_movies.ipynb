```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
path = '/content/drive/MyDrive/imdb_top_1000.csv'
df = pd.read_csv(path)
df.head()
```

```python
df.info()
```

```python
pd.options.display.float_format = '{:,.2f}'.format
df.describe(include='all')
```

# ***Data Cleaning / Preprocessing***

## **Rename Columns**

```python
df.columns = df.columns.str.lower()
df.columns
```

## **Drop Some Columns**

```python
df = df.drop(['poster_link', 'overview'], axis=1)
df.columns
```

## **Duplicate Values**

```python
#Check duplicates
print("Number of duplicate rows:")
df.duplicated().sum()
```

## **Missing Values**

```python
print("Number of missing values:")
df.isnull().sum()
```

```python
round(df.isna().mean()*100,2)
```

```python
df['gross']= df['gross'].replace(',','', regex=True)
df['gross'] = pd.to_numeric(df['gross'], errors='coerce')
df['gross'].dtype
```

```python
df['released_year'] = pd.to_numeric(df['released_year'], errors='coerce')
df['released_year'].dtype
```

```python
df['runtime'] = df['runtime'].replace('min','', regex=True)
df['runtime'] = pd.to_numeric(df['runtime'], errors='coerce')
df['released_year'].dtype
```


#**Fill missing values**

```python
df['certificate'] = df['certificate'].fillna(0)
df['gross'] = df['gross'].fillna(df['gross'].mean())
df['meta_score'] = df['meta_score'].fillna(df['meta_score'].mean())
df['released_year']= df['released_year'].fillna(df['released_year'].mean())
df.isnull().sum()
```

### ***Distribution of certificate***

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='certificate', data=df, order=df['certificate'].value_counts().index, palette= 'Set1')
plt.title("Distribution of Certificates")
plt.xticks(rotation=45)
plt.show()
```

```python
top_10_rating = df.sort_values(by='imdb_rating', ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='series_title', y='imdb_rating', hue='released_year', data=top_10_rating, palette='viridis', s=100)
plt.xticks(rotation=90)
plt.title('Top 10 Rating Movies by IMDB Rating and Released Year')
plt.xlabel('Movie Title')
plt.ylabel('IMDB Rating')
plt.tight_layout()
plt.show()
```

***Insight:***

The top 10 highest-rated movies span across multiple decades (1960s2010s), showing that great movies are not limited to a specific era. However, the highest-rated one belongs to 2020, reflecting that recent productions can still achieve exceptional audience and critic appreciation.  

***Recommendation:***

Studios should preserve classic storytelling values while also embracing modern production quality, as both old and new movies continue to gain top ratings.

# ***Relationships & Correlations:***

# **Correlation heatmap**


```python
plt.figure(figsize=(12, 5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

 ***Correlation Insights & Recommendations Summary***
1. **Gross vs IMDb Rating**

***Insight:***
There is a weak positive correlation (~0.10) between IMDb rating and gross earnings, indicating that critically acclaimed movies dont always achieve higher box office revenue.

***Recommendation:***
Studios should not rely solely on high ratings; instead, invest in marketing, release timing, and franchise development to ensure financial success.

**2. Gross vs Runtime**

***Insight:***
A weak correlation (~0.13) shows that movie length has minimal influence on box office performance.

***Recommendation:***
Focus on storytelling and pacing rather than extending runtime. Audience engagement matters more than duration.

***3. Gross vs Released Year***

***Insight:***
Profits have generally increased over time, especially after 2000, due to global distribution and digital cinema expansion.

***Recommendation:***
Studios should prioritize new releases with international reach to maximize earnings.

**4. IMDb Rating vs Meta Score**

***Insight:***
Positive correlation (~0.27) indicates general agreement between critics (Meta Score) and audiences (IMDb rating), though some exceptions exist.

Recommendation:
Balance between critical quality and audience satisfaction to ensure both prestige and sustained popularity.

**5. Runtime vs IMDb Rating**

***Insight:***
Movies with moderate runtimes (120150 minutes) tend to have slightly higher ratings.

***Recommendation:***
Aim for balanced runtimes that allow story depth without viewer fatigue.

**6. No. of Votes vs Gross**

***Insight:***
Strong correlation (~0.57) reveals that popularity and visibility drive financial success.

***Recommendation:***
Increase marketing and global distribution to attract larger audiences and higher vote counts.

**7. No. of Votes vs IMDb Rating**

***Insight:***
Moderate-to-strong correlation (~0.48) suggests that higher-rated movies attract more audience engagement and votes.

***Recommendation:***
Promote highly rated titles widely; popularity reinforces positive feedback loops and revenue.

**8. Meta Score vs Gross**

***Insight:***
Very weak or near-zero correlation (~-0.017) implies that critical reviews alone do not guarantee financial success.

***Recommendation:***
Balance critical acclaim with audience appeal; combine artistic excellence with commercial strategies.

# ***Average Rating by Genre***

```python
average_rating_by_genre = df.groupby('genre')['imdb_rating'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x=average_rating_by_genre.values, y=average_rating_by_genre.index, palette='Set2')
plt.title('Average IMDB Rating by Genre')
plt.xlabel('Average IMDB Rating')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

**Insight:**  
Genres with the highest average ratings are **War (8.01), Western (8.00), and Film-Noir (7.99)**, showing that critically acclaimed movies often belong to these categories.  

**Recommendation:**  
Studios can invest more in underrepresented but high-rated genres (e.g., War, Western, Film-Noir) to attract critical acclaim and long-term audience appreciation.

### **Director vs Average IMDB Rating (Top Directors)**


```python
top_directors = df['director'].value_counts().head(10).index
avg_rating_by_director = df[df['director'].isin(top_directors)].groupby('director')['imdb_rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_rating_by_director.values, y=avg_rating_by_director.index, palette="coolwarm")
plt.title("Average IMDB Rating of Top 10 Directors")
plt.xlabel("Average IMDB Rating")
plt.ylabel("Director")
plt.show()
```

***Insight:***

Among the most prolific directors, some consistently deliver higher-rated movies. This highlights the influence of director style and reputation on critical reception.

***Recommendation:***

Studios can prioritize collaborations with directors who consistently achieve high ratings, as they not only bring quality but also attract loyal audiences.
